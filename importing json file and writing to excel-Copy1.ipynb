```python
#importing json file 
import json
with open('geocodinglogfle.json', 'r') as f:
    json_list = json.load(f)
```

```python
len(json_list)
```

```python
#creating a pandas dataframe with the required columns
import pandas as pd
data = pd.DataFrame(columns=['api_formatted_address','api_lat','api_lng','location_type','api_json_resp'])
```

```python
#creating a set to hold exceptions and setting the range for loop
z=len(json_list)
y =set()
m = z-1
m
```

```python
#getting the required data traversing through the loop
for i in range(m):
    try:
        x=json_list[i]
        obj1 = x['results']
        obj2 = obj1[0]['address_components']
        formatted_address = obj1[0]['formatted_address']
        obj3 = obj1[0]['geometry']
        location_type = obj3['location_type']
        obj4 = obj3['location']
        latitude = obj4['lat']
        longitude = obj4['lng']
        data = data.append({'api_formatted_address':formatted_address,'api_lat':latitude,'api_lng':longitude,'location_type':location_type,'api_json_resp':obj1},ignore_index=True)
    except:
        y = json_list[i]
```

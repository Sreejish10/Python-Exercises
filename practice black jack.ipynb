```python
# getting rank suits and values for the deck
import random

suits = ('Hearts','Diamonds','Spades','Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}

playing = True
```

```python
class Card():
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        
    def __str__(self):
        return self.rank + ' of ' +self.suit
```

```python
#deck class

class Deck():
    
    def __init__(self):
        self.deck= []
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))
    def __str__(self):
        deck_comp = ''
        for card in self.deck:
            deck_comp += '\n'+card.__str__()
        return 'The deck has: ' + deck_comp
        
        
    # we need a function to shuffle 52 deck card
    def shuffle(self):
        random.shuffle(self.deck)
    
    def deal(self):
        single_card = self.deck.pop()
        return single_card
```

```python
#we create a hand class which gives idea of the cards one is holding and along with that it gives the sum value of the cards

class Hand():
    def __init__(self):
        self.card = [] #starts with a empty list as in deck
        self.value = 0 #starts with zero value
        self.aces = 0
        
    def add_card(self,card):
        #card passed in is from class Deck.deal
        # which return single card
        self.card.append(card)
        self.value += values[card.rank]
        
        #keep track of aces
        if card.rank== 'Ace':
            self.aces +=1
        
    def adjust_for_ace(self):
        # here we need to check if we have self.value >21 
        # then change Ace value from 11 to 1
        if self.value > 21 and self.aces > 0: #(if self.value >21 and self.aces), #both is same as self.aces acts as boolean)
            self.value -=10
            self.aces -=1
        
```

```python
# now we create a chip class

class Chips():
    def __init__(self,total=100):
        self.total = total
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
        
    def lose_bet(self):
        self.total -= self.bet
```

```python
# create a function to create bet

def take_bet(chips):
    
    while True:
        try:
            chips.bet = int(input("HOw many chips would you like to bet ? "))
        except:
            print("\n Sorry, please enter an integere value ")
        else:
            if chips.bet > chips.total:
                print(" Sorry, you donot have suffiecient chips, you only have {}".format(chips.total))
            else:
                break
```

```python
# create a function take hits

def hit(deck,hand):
    hand.add_card(deck.deal())
    hand.adjust_for_ace()
```

```python
#Write a function prompting the Player to Hit or Stand
def hit_or_stand(deck,hand):
    global playing # to control for upcoming while loop
    
    while True:
        x = input(" Would like to take a hit or stand, enter 'h' or 's' : ")
        
        if x[0].lower() == 'h' :
            hit(deck,hand)
        
        elif x[0].lower() == 's':
            print("Player stands, dealer is playing")
            playing = False
            
        else:
            print("Sorry please try again")
            continue
        break
```
